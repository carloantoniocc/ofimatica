```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from itertools import cycle
from sqlalchemy import create_engine
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
plt.plot(x,y)
plt.show
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.hist(y)
plt.show
```

```python
## figure es un lienzo donde poner axes
fig = plt.figure()
## crear la figura parametros 0.5 corresponde al ancho y 0.9 corresponde a largo
axes = fig.add_axes([0.1,0.1,0.5,0.9])
axes.plot(x,y,'b')
```

```python
## metodo orientado a objetos (mas perzonalizable) 
fig = plt.figure()
## crear la figura parametros 0.5 corresponde al ancho y 0.9 corresponde a largo
axes = fig.add_axes([0.1,0.1,0.8,0.9])
axes2 = fig.add_axes([0.2,0.55,0.4,0.3])

axes.plot(x,y,'b')
axes2.plot(y,x,'b')
axes2.set_facecolor('gray')

fig.show()
```

```python
x=np.linspace(0,5,11)
y=np.sin(x)
```

```python
#subplots
fig, axes = plt.subplots(2,4)
axes[0,0].plot(x,np.cos(x))
axes[0,1].plot(x,np.sin(x),'r')
axes[0,2].plot(x,np.tan(x),'y')
axes[0,3].plot(x,np.cos(x)**2)
fig.tight_layout()
```

```python
## titulos leyendas
## figsize=(5,5) modifica el tamao
fig , axes = plt.subplots(1,2,figsize=(5,5))
axes[0].plot(x,y,label="$sin(x)$")
axes[0].set_title('relacion X - Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend()
axes[1].plot(y,x,'r',label="$cos(x)$")
axes[1].set_title('relacion Y - x')
axes[1].set_xlabel('X')
axes[1].set_ylabel('Y')
axes[1].legend(loc='lower left')
```
